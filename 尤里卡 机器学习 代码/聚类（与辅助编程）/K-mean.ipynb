```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score, normalized_mutual_info_score
import pandas as pd

# 1. 加载数据
iris = load_iris()
X = iris.data
y_true = iris.target

# 2. 聚类建模
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# 3. 计算聚类性能指标
# 轮廓系数 (Silhouette Score) - 衡量聚类的紧密度和分离度
silhouette_avg = silhouette_score(X, labels)
print(f"轮廓系数 (Silhouette Score): {silhouette_avg:.4f}")
print("   - 范围: [-1, 1], 越接近1表示聚类效果越好")

# 调整兰德指数 (Adjusted Rand Index) - 衡量聚类结果与真实标签的一致性
ari = adjusted_rand_score(y_true, labels)
print(f"调整兰德指数 (ARI): {ari:.4f}")
print("   - 范围: [-1, 1], 越接近1表示与真实标签越一致")

# 标准化互信息 (Normalized Mutual Information) - 衡量聚类结果的信息量
nmi = normalized_mutual_info_score(y_true, labels)
print(f"标准化互信息 (NMI): {nmi:.4f}")
print("   - 范围: [0, 1], 越接近1表示聚类效果越好")

# 4. 构建结果表
df = pd.DataFrame(X)
df["cluster"] = labels
df["true_label"] = y_true
print("\n聚类结果表:")
print(df)
```
