```python
'''
使用 inspection.DecisionBoundaryDisplay 显示决策边界

可视化分类器的决策边界：将分类器在特征空间中划分不同类别的边界显示出来。这有助于理解模型如何对不同输入做出分类。
'''
import matplotlib.pyplot as plt  # 导入绘图库matplotlib的pyplot模块，用于绘图
import numpy as np  # 导入数值计算库numpy
from sklearn.datasets import load_iris  # 从sklearn导入鸢尾花数据集加载函数
from sklearn.inspection import DecisionBoundaryDisplay  # 导入决策边界可视化工具
from sklearn.tree import DecisionTreeClassifier  # 导入决策树分类器

iris = load_iris()  # 加载鸢尾花数据集
feature_1, feature_2 = np.meshgrid(
    np.linspace(iris.data[:, 0].min(), iris.data[:, 0].max()),  # 生成第一个特征的网格点
    np.linspace(iris.data[:, 1].min(), iris.data[:, 1].max())   # 生成第二个特征的网格点
)  # meshgrid生成二维网格，用于绘制决策边界

grid = np.vstack([feature_1.ravel(), feature_2.ravel()]).T  # 将网格点展开并组合成二维坐标点数组

tree = DecisionTreeClassifier().fit(iris.data[:, :2], iris.target)  # 用前两个特征训练决策树分类器

y_pred = np.reshape(tree.predict(grid), feature_1.shape)  # 对网格点进行预测，并重塑为网格形状

display = DecisionBoundaryDisplay(
    xx0=feature_1, xx1=feature_2, response=y_pred  # 创建决策边界显示对象
)
display.plot()  # 绘制决策边界

display.ax_.scatter(
    iris.data[:, 0], iris.data[:, 1], c=iris.target, edgecolor="black"  # 在图上绘制原始数据点
)

plt.show()  # 显示图形窗口
plt.savefig('分类结果.png', dpi=300, bbox_inches='tight')  # 保存图片到当前目录
```
