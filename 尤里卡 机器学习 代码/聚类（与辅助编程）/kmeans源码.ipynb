```python
import random  # 导入 random，用来随机选择初始聚类中心
import math    # 导入 math，用来计算平方根
```

```python


def euclidean_distance(point1, point2):  # 定义函数：计算两个点之间的欧氏距离
    distance = 0                         # 初始化距离平方和为 0

    for i in range(len(point1)):         # 遍历点中的每一个维度
        distance += (point1[i] - point2[i]) ** 2  # 累加每个维度差值的平方

    return math.sqrt(distance)           # 对平方和开根号，得到欧氏距离


def assign_clusters(dataset, centroids): # 定义函数：把每个样本分配到最近的中心点
    clusters = [[] for _ in centroids]   # 创建若干个空簇，每个中心点对应一个簇

    for point in dataset:                # 遍历数据集中的每一个样本点
        distances = []                   # 用来保存当前点到所有中心点的距离

        for centroid in centroids:       # 遍历每一个聚类中心
            dist = euclidean_distance(point, centroid)  # 计算当前点到该中心点的距离
            distances.append(dist)       # 把距离加入 distances 列表

        nearest_index = distances.index(min(distances)) # 找到距离最小的中心点索引
        clusters[nearest_index].append(point)           # 把当前点加入最近中心点对应的簇

    return clusters                      # 返回分好类的所有簇


def calculate_centroid(cluster):         # 定义函数：计算一个簇的新中心点
    num_points = len(cluster)            # 计算这个簇里面有多少个样本点
    num_features = len(cluster[0])       # 计算每个样本点有多少个特征维度

    centroid = []                        # 用来保存新的中心点

    for i in range(num_features):        # 遍历每一个特征维度
        feature_sum = 0                  # 初始化当前维度的总和为 0

        for point in cluster:            # 遍历簇中的每一个样本点
            feature_sum += point[i]      # 累加当前维度的值

        centroid.append(feature_sum / num_points) # 当前维度平均值作为中心点该维度的值

    return centroid                      # 返回新的中心点


def update_centroids(clusters, old_centroids): # 定义函数：根据当前簇更新所有中心点
    new_centroids = []                   # 用来保存更新后的中心点

    for i in range(len(clusters)):       # 遍历每一个簇
        cluster = clusters[i]            # 取出当前簇

        if len(cluster) == 0:            # 如果当前簇是空的
            new_centroids.append(old_centroids[i]) # 保留原来的中心点，避免报错
        else:                            # 如果当前簇不是空的
            new_centroid = calculate_centroid(cluster) # 计算当前簇的新中心点
            new_centroids.append(new_centroid)         # 把新中心点加入列表

    return new_centroids                 # 返回所有更新后的中心点


def centroids_changed(old_centroids, new_centroids): # 定义函数：判断中心点是否发生变化
    for i in range(len(old_centroids)):  # 遍历每一个中心点
        if old_centroids[i] != new_centroids[i]: # 如果有任意一个中心点不同
            return True                  # 说明中心点发生了变化

    return False                         # 所有中心点都没变，返回 False




```

```python
def k_means(dataset, k, max_iterations=100): # 定义 K-means 主函数
    centroids = random.sample(dataset, k)    # 从数据集中随机选 k 个点作为初始中心点

    for iteration in range(max_iterations):  # 最多迭代 max_iterations 次
        clusters = assign_clusters(dataset, centroids) # 把样本分配到最近的中心点

        new_centroids = update_centroids(clusters, centroids) # 根据簇更新中心点

        if not centroids_changed(centroids, new_centroids):   # 如果中心点不再变化
            break                         # 停止迭代

        centroids = new_centroids         # 用新的中心点替换旧的中心点

    return clusters, centroids            # 返回最终的簇和最终中心点

```

```python
dataset = [                               # 定义一个二维数据集
    [1, 2],                               # 第 1 个样本点
    [1, 4],                               # 第 2 个样本点
    [1, 0],                               # 第 3 个样本点
    [10, 2],                              # 第 4 个样本点
    [10, 4],                              # 第 5 个样本点
    [10, 0]                               # 第 6 个样本点
]

k = 2                                     # 设置要分成 2 个簇
```

```python
clusters, centroids = k_means(dataset, k) # 调用 K-means 算法进行聚类

print("最终聚类结果：")                    # 打印提示文字
for i, cluster in enumerate(clusters):    # 遍历每一个簇
    print(f"第 {i + 1} 类:", cluster)      # 打印当前簇中的样本点

print("最终中心点：", centroids)           # 打印最终的聚类中心
```
