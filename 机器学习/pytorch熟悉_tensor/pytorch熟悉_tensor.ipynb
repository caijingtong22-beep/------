## chunck

```python
B = torch.randn(5, 2) # 生成一个5行2列的二维张量，元素服从均值为0，标准差为1的正态分布

# 想分成 3 块
res = torch.chunk(B, chunks=3, dim=0)

print(len(res))       # 结果还是 3 块
print(res[0].shape)   # torch.Size([2, 2])
print(res[1].shape)   # torch.Size([2, 2])
print(res[2].shape)   # torch.Size([1, 2]) -> 最后一块补齐剩下的
print(B)         
```
